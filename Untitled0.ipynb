```python
#import libraries 
import numpy as np 
import matplotlib.pylab as plt
import librosa 
import librosa.display

#import audio file 
!git clone https://github.com/jcdevaney/imc2023.git
```

```python
# open audio file avm2.wav with librosa.load()
file = 'imc2023/audioFiles/avm.wav'
samplingRate = 44100
sig, sr = librosa.load(file,mono=True,sr=samplingRate)
```

```python
plt.plot(sig)
```

```python
#plot time domain representation of the signal 
time=np.arange(0,sig.shape[0])/sr
plt.plot(time,sig)

```

```python
#add x and y labels
plt.plot(time,sig)
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
```
